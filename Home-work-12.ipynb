```python
from sympy import *
from sympy.plotting import plot
init_printing()
x = Symbol('x')
```

```python
f = x**3 - 50 * x
g = -x**4 + 88 * x**2 - 241
```

```python
#1. Найти координаты точек пересечения
point_x = solve(f - g, x)

point_y = list()
for i in range(len(point_x)):
    point_y.append(f.subs(x, point_x[i]))
    print(f'Точка {i+1}\nx: {point_x[i]} \ny: {point_y[i]}')
#point_x
#point_y
```

```python
#2. Построить графики функций в одной системе координат
plot(f, g)
```

```python
#3. Построить график функции пересечения
h = f - g
plot(h)
```
